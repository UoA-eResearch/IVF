```python
from PIL import Image, ImageFile # Image manipulation
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
from tqdm.auto import tqdm
files = glob("M776_S0112/*.jpg")
len(files)
```

```python
image = np.array(Image.open(files[10]))

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10,10), sharex=True, sharey=True)

ax1.imshow(image)
ax2.imshow(image[:, :, 0], cmap="Reds_r")
ax3.imshow(image[:, :, 1], cmap="Greens_r")
ax4.imshow(image[:, :, 2], cmap="Blues_r")
plt.tight_layout()
```
